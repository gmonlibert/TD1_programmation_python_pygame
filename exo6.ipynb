```python
# -*- coding: utf-8 -*-

# definitions de quelques couleurs ... par synthèse additive des sous pixels 
# rouge, vert et bleu.: les intesnsités des sous-pixels varient entre 0 et 255.
# il y a 256 valeurs différentes car chaque intensité est codée sur 8 bits
# (2^8 = 256)

rouge = [255,0,0]
vert = [0,255,0]
bleu = [0,0,255]
jaune = [255,255,0]
magenta = [255,0,255]
cyan = [0,255,255]
blanc = [255,255,255]
orange = [127,255,0]

# importation du module graphique 2D pygame
import pygame

# initialisation du module d'affichage 2D pygame
pygame.display.init()

# Création de la fenêtre graphique de largeur 900 et hauteur 600
size = [900,600]
screen = pygame.display.set_mode(size)
# affichage et mise à jour de la fenêtre 2D
pygame.display.flip()

""" DEBUT DU CODE ICI ... """

debut = [450,0]
fin = [450,600]
pygame.draw.line(screen, blanc, debut, fin, 1)
pygame.display.flip()

""" FIN DU CODE ICI ... """

# boucle pour faire une pause en attente de l'appui sur la touche echap ou 
# sur le bouton de fermeture de la fenêtre
continuer = 1
while continuer == 1:
    for event in pygame.event.get():
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_ESCAPE:
                continuer = 0
        if event.type == pygame.QUIT:
            continuer = 0
        if event.type == pygame.MOUSEBUTTONDOWN:
            clic = pygame.mouse.get_pos()
            sym = [900 - clic[0], clic[1]]
            if clic[0] < 450 :
                couleur = bleu
            else :
                couleur = jaune
            pygame.draw.circle(screen, couleur, sym, 50, 3)
            pygame.display.flip()
            
# sortie du programme
pygame.quit()
```
